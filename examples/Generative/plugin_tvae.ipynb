```python
# stdlib
import warnings

warnings.filterwarnings("ignore")

# # third party
# from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np
# synthcity absolute
from synthcity.plugins import Plugins
```

### Load dataset


```python
data = pd.read_csv("/content/df_train_AG-50.csv")
X = data.drop("label", axis = 1)
y = data["label"]
```

```python
# synthcity absolute
from synthcity.plugins.core.dataloader import GenericDataLoader

# X, y = load_diabetes(return_X_y=True, as_frame=True)
# X["target"] = y

loader = GenericDataLoader(data, target_column="label", sensitive_columns=[])

loader.dataframe()
```

### Train the generator


```python
# synthcity absolute
from synthcity.plugins import Plugins

# define the model hyper-parameters
plugin_params = dict(
    n_iter = 1000,  # epochs
    lr = 0.0002,
    weight_decay = 1e-4,
    batch_size = 1000,
)
eval_plugin = "tvae"
syn_model = Plugins().get(eval_plugin, **plugin_params)

syn_model.fit(loader)
```

### Generate new samples


```python
test = pd.read_csv("/content/df_test_AG-50.csv")
```

```python
syn_model.generate(count=10).dataframe()
```

```python
syn = syn_model.generate(count=test.shape[0])
```

```python
syn.dataframe().to_csv("tvae.csv", index=False)
```

```python
test.to_csv("real.csv", index=False)
```

### Benchmarks

```python
# synthcity absolute
from synthcity.benchmark import Benchmarks

score = Benchmarks.evaluate(
    [
        (eval_plugin, eval_plugin, {"n_iter": 50})
    ],  # (testname, plugin, plugin_args). REPLACE {"n_iter" : 50} with {} for better performance
    loader,
    repeats=2,
    metrics={"detection": ["detection_mlp"]},  # DELETE THIS LINE FOR ALL METRICS
)
```

```python
Benchmarks.print(score)
```
